```python
import numpy as np, scipy as sp, keras, tensorflow, re, os, glob, pandas as pd
from keras import backend as K
import statsmodels
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
from scipy import stats
from sklearn import preprocessing
```

```python
df=pd.read_csv("./trainingData.csv",low_memory=False, nrows=5001)
```

```python
df['sex'] = df['sex'].astype('category')
df['social_class'] = df['social_class'].astype('category')
df['primary_business'] = df['primary_business'].astype('category')
df['secondary_business'] = df['secondary_business'].astype('category')
df['type_of_house'] = df['type_of_house'].astype('category')
df['loan_purpose'] = df['loan_purpose'].astype('category')

cat_columns = df.select_dtypes(['category']).columns
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
```

```python
def qqplot(x, y, **kwargs):
    _, xr = stats.probplot(x, fit=False)
    _, yr = stats.probplot(y, fit=False)
    plt.scatter(xr, yr, **kwargs)
```
